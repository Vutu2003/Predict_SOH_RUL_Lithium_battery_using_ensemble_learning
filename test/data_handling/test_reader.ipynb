```python
import sys

sys.path.append('../../src/data_handling')
```

```python
from reader import BatteryDataReader

# 1. Xác định ID pin và đường dẫn thư mục
battery_to_load = "B0005"

data_directory = "../../dataset/" 

# 2. Khởi tạo đối tượng BatteryDataReader
print(f"\nInitializing reader for battery: {battery_to_load}")
b0005_reader = BatteryDataReader(battery_id=battery_to_load, data_dir=data_directory)

# 3. Tải dữ liệu
print("\nAttempting to load data...")
load_status = b0005_reader.load_data()

# 4. Kiểm tra và sử dụng dữ liệu nếu tải thành công
if load_status:
    print("\nData for B0005 loaded successfully.")

    # Lấy các DataFrame
    df_capacity_b0005, df_charge_b0005, df_discharge_b0005 = b0005_reader.get_raw_data()

    # Hiển thị thông tin cơ bản của từng DataFrame để xác nhận
    print("\n--- Capacity Data (B0005) ---")
    if df_capacity_b0005 is not None:
        print(df_capacity_b0005.info())
        print(df_capacity_b0005.head())
    else:
        print("Capacity DataFrame is None.")

    print("\n--- Charge Data (B0005) ---")
    if df_charge_b0005 is not None:
        print(df_charge_b0005.info())
        print(df_charge_b0005.head())
    else:
        print("Charge DataFrame is None.")

    print("\n--- Discharge Data (B0005) ---")
    if df_discharge_b0005 is not None:
        print(df_discharge_b0005.info())
        print(df_discharge_b0005.head())
    else:
        print("Discharge DataFrame is None.")
else:
    print(f"\nFailed to load complete data for {battery_to_load}. Please check file paths and file integrity in '{data_directory}'.")
```
