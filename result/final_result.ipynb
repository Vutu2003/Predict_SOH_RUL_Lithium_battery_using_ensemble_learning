# COMPARE RMSE 

```python
# With split data
rmse_soh_b0005_split = np.array([0.8593, 1.0012, 1.0254])
rmse_soh_b0006_split = np.array([1.3121, 1.0558, 1.2773])
rmse_soh_b0007_split = np.array([0.9847, 1.0906, 1.0042])
rmse_soh_b0018_split = np.array([1.2212, 1.0739, 1.1046])
rmse_rul_b0005_split = np.array([4.4518, 5.9502, 6.1709])
rmse_rul_b0006_split = np.array([3.1520, 3.0820, 2.9893])
rmse_rul_b0007_split = np.array([5.2466, 6.2059, 6.7202])
rmse_rul_b0018_split = np.array([7.4488, 6.7090, 7.1594])
```

```python
# soh
# --- 2. Chuẩn bị dữ liệu để vẽ đồ thị ---
algorithms = ['Random Forest', 'XGBoost', 'Stacking']
batteries = ['B0005', 'B0006', 'B0007', 'B0018']

# Tạo một danh sách để chứa dữ liệu theo cấu trúc "dài" (long format)
plot_data = []

# Dữ liệu cho pin B0005
for i, algo in enumerate(algorithms):
    plot_data.append({'Algorithm': algo, 'Battery': 'B0005', 'RMSE (%)': rmse_soh_b0005_split[i]})

# Dữ liệu cho pin B0006
for i, algo in enumerate(algorithms):
    plot_data.append({'Algorithm': algo, 'Battery': 'B0006', 'RMSE (%)': rmse_soh_b0006_split[i]})

# Dữ liệu cho pin B0007
for i, algo in enumerate(algorithms):
    plot_data.append({'Algorithm': algo, 'Battery': 'B0007', 'RMSE (%)': rmse_soh_b0007_split[i]})

# Dữ liệu cho pin B0018
for i, algo in enumerate(algorithms):
    plot_data.append({'Algorithm': algo, 'Battery': 'B0018', 'RMSE (%)': rmse_soh_b0018_split[i]})

# Chuyển thành một DataFrame của Pandas
df_plot = pd.DataFrame(plot_data)


# --- 3. Vẽ biểu đồ cột 2D được nhóm ---
plt.style.use('seaborn-v0_8-whitegrid') # Sử dụng một style cho đẹp hơn
fig, ax = plt.subplots(figsize=(12, 7)) # Tạo một figure và axes

# Sử dụng seaborn để vẽ barplot
sns.barplot(data=df_plot, x='Algorithm', y='RMSE (%)', hue='Battery', ax=ax, palette='viridis')

# Thêm các nhãn và tiêu đề
ax.set_title('So sánh Sai số RMSE của các Mô hình trên từng Pin với cách chia tỉ lệ 3:1 (SOH)', fontsize=16, fontweight='bold')
ax.set_xlabel('Thuật toán', fontsize=12, fontweight='bold')
ax.set_ylabel('RMSE (%)', fontsize=12, fontweight='bold')
ax.legend(title='Pin Kiểm thử', title_fontsize='11')

# Thêm giá trị lên trên mỗi cột để dễ đọc
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=10, padding=3)

# Tự động điều chỉnh layout và lưu hình
plt.tight_layout()
#plt.savefig('comparison_rmse_split_data.png', dpi=300)

print("Đồ thị đã được tạo và lưu với tên 'comparison_rmse_split_data.png'")
```
